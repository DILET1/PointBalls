# The "Classic Runge-Kutta Method" a.k.a "RK4"

The Runge-Kutta methods are a family of iterative methods to solve simultaneous nonlinear equations. The Crank-Nicolson FDM I used in another project, [SchrodingerSolver](https://github.com/DILET1/SchrodingerSolver), is one example of an implicit Runge-Kutta method.

The RK4 method is formulated as follows:

We have a function, $y(t)$, with initial conditions $y_0$ and $t_0$. We are given a function $f(y,t)$, which satisfies

$$\frac{dy}{dt} = f(y,t),\quad y(t_0)=y_0$$

The objective is to find how y evolves with time. This is done by iteratively calculating $y$. We define a time stepping $L_t$, define $t_n$ as $t_0 + nL_t$ and define $y_n$ as the $n$th iterative solution to the question, corresponding to $y(t_n)$. $y_{n+1}$ is then defined as:

$$y_{n+1} = y_n + \frac{L_t}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$
$k_1, k_2, k_3$ and $k_4$ are defined as below:
$$k_1 = f(y_n, t_n)$$
$$k_2 = f(y_n + k_1\frac{L_t}{2}, t_n + \frac{L_t}{2})$$
$$k_3 = f(y_n + k_2\frac{L_t}{2}, t_n + \frac{L_t}{2})$$
$$k_4 = f(y_n + k_3L_t, t_n + L_t)$$

If we examine the second term in the expression for $y_n+1$, we can see that it's $L_t$ multiplied by a weighted average of the $k_i$ with more weight towards $k_2$ and $k_3$.

$k_1$ is just the rate of change of $y$ at $t_n$.

After calculating $k_1$, we use it to find the point at which to calculate $k_2$, in the middle of the time stepping interval.

$k_3$ uses the same formula as $k_2$, but replaces the dependence on $k_1$ with $k_2$.

$k_4$ uses $k_3$ to approximate where the endpoint of the interval for $y$ is, and calculates the rate of change there.

In effect, we're calculating four slopes at different points in the interval, and averaging them to approximate what $y_{n+1}$ is.
