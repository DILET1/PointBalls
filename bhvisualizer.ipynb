# The Barnes-Hut N-body Simulation Algorithm

For very large N, the asymptotic complexity of simulating the system is $O(N^2)$. This can very quickly get too slow or expensive to simulate. The Barnes-Hut approximation method has an asymptotic complexity of $O(NlogN)$, significantly more manageable. As I learned in USACO, $O(NlogN)$ is good enough, $O(N^2)$ is bad.

At its core, the Barnes-Hut method is essentially just utilizing the realization that once sufficiently far away, a system of particles can just be approximated a single particle moving about the system's center of mass.

To realize this, space is recursively divided into quadrants(2D) or octants(3D). We will be working in 2D for now. The resultant quadrants are subdivided further until every region of space has either 0 or 1 body within it. Then, we use a tree representation to store the current system state - each node corresponds to a quadrant, with 4 children representing its daughter quadrants (if any exist). The node contains the data that approximates its 4 children as a single particle moving about the COM of the system.

We introduce a parameter $\theta = \frac{s}{r}$ where $s$ is the sidelength of the quadrant and $r$ is the distance between the COM of the quadrant and the point mass whose motion we're trying to calculate. For a sufficiently small $\theta$, we can treat the quadrant as a point mass. Otherwise, we must treat each daughter quadrant individually. This can be done with a Depth-First Search approach in $O(NlogN)$

## Calculation process for mass $m_1$

We start at the root of the tree. We will use a DFS tree traversal strategy, with the following procedure for each node:

1. Determine if $m_1$ is included in the quadrant represented in this node, or if $\theta$ for the node is larger than our allowed threshold ($\theta$ = 0 for leaf nodes). If either are true, call DFS on each daughter quadrant. This step effectively checks if we cannot treat the current quadrant as a single point mass. Note that if the current node only contains $m_1$, we'll just return 0 as obviously $m_1$ doesn't contribute to itself.
2. Since we have determined it is allowable for the quadrant to be treated by its COM representation, we calculate its contributions to the motion of $m_1$
3. return.

This whole process has an asymptotic complexity of $O(NlogN)$.

In addition, every time stepping, for each node we must calculate its center of mass based on its daughter quadrants (besides leaf nodes). This has asymptotic complexity $O(N)$.
